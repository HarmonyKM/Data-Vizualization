## **Load the dataset from data.world**

```python
df = pd.read_excel('https://query.data.world/s/n2joizfinrk2n5fjn7yuavcj4z4bte')
df
```

```python
# Clean up Nintendo Switch Platform by making it 2 columns
df[['Type','Country']] = df['Nintendo Switch Platform'].str.split('/',expand=True)
df
```

```python
# Simplify Type column to just hardware or software
df['Type'] = df['Type'].str.replace('Nintendo Switch ','')
df
```

```python
## Drop original columns
plot_df = df.drop(columns=['Nintendo Switch Platform','Metric'])
plot_df
```

```python
## Set index for plotting
plot_df = plot_df.set_index(['Type','Country'])
plot_df
```

```python
## Separate hardware vs software 
software = plot_df.loc['software'].T

hardware = plot_df.loc['hardware'].T
display(software,hardware)
```

```python
## Using specific colors for each country
color_map = {'Japan':'darkblue','The Americas':'darkgreen',
          "Europe":'orange','Other':'darkred'}
colors = [color_map[c] for c in hardware.columns]
colors
```

## **Reporoducing Plot**

```python
# Plot with twin axes
fig, ax1 = plt.subplots(figsize=(10,5))
ax2 = ax1.twinx()
hardware.plot(kind='bar',ax=ax1,stacked=True,color=colors,rot=0,alpha=0.5)
ax1.set_ylabel('Hardware Sold (in 10k units)')

software.plot(marker='o',ax=ax2, color=colors)
ax2.set_ylabel('Software Sold (in 10k units)')
ax1.legend()
ax2.legend()

ax1.set_title('Nintendo Switch Sales');
```

## **Improving the Plot - Subplots**

```python
with plt.style.context('ggplot'):
    ## plot as subplots
    fig, axes = plt.subplots(nrows=2,sharex=True,figsize=(8,6))

    ## Plot hardware 
    hardware.plot(kind='bar',stacked=True,color=colors,ax=axes[0])#,title='Hardware Sales')
    axes[0].set_ylabel('Hardware Sales (in 10Ks units)')
    axes[0].set_title('Nintendo Switch Sales')
    axes[0].legend()

    # Plot software
    software.plot(kind='line',marker='o',color=colors,ax=axes[1])
    axes[1].set(ylabel='Software Sales (in 10Ks units',
                xlabel='Month');
    axes[1].legend()

    ## Fix overlap
    plt.tight_layout()
```

## **Task 2**

- Dataset chosen: [2020/W46: Majority of Advertising Dollars Now Spent Online](https://data.world/makeovermonday/2020w46)

<img src="https://media.data.world/4nPsFb3CSJSXMGmxgBbH_WhatsApp%20Image%202020-11-14%20at%2018.19.31.jpeg">

```python
## Read in data from data.world
df = pd.read_excel("https://query.data.world/s/dzdxamcuc5khsneejpodk56moucd4s")
df
```

```python
# How many unique values in Type and Source columns?
df.nunique()
```

```python
# Displaying unique values in Source column to standardize
df['Source'].unique()
```

```python
# The lines plotted in example are not as simple as either the Type or Source columns

# make a dict map for new/final names for plot
source_map = {'National TV': np.nan,
            'Local TV':np.nan,
            'Total TV':'Television',
            'Total Radio':'Radio',
            'Total Newspapers':"Newspapers",
            'Total Magazines':'Magazines',
            'Pure-Play Internet': 'Internet',
            'Search':np.nan,
            'Ex-Search':'Internet',
            'Media total': np.nan,
            }
source_map
```

```python
# Make new source name column
df['Source Name'] = df['Source'].replace(source_map)
df
```

```python
# Get single line for each type
plot_df = df.dropna().groupby("Source Name").sum()
plot_df
```

```python
## Convert to %
plot_df_perc = plot_df / plot_df.sum()  * 100
plot_df_perc
```

```python
## Get unique type names to plot 
types = plot_df_perc.index.to_list()
types
```

```python
## For each type, save the years to plot and values
data = {}
for type_ in types:
    data[type_] = plot_df_perc.loc[type_].values

data
```

```python
## Save list of columns to plot
years = plot_df_perc.columns
years
```

```python
## Reproduce line plot
with plt.style.context('ggplot'):
    ## Loop through dict to plot each type
    plt.figure(figsize=(10,6))
    for type_, values in data.items():
        plt.plot(years,values, marker='o',label=type_)

    ## Add legend 
    plt.legend(loc='upper left')
    plt.xlim(2012,2024)

    ## Get rid of frame 
    plt.box(False)
    # plt.yticks([]);

    fig = plt.gcf()
```

```python
## Saving values to annotate at end of plot
vals_to_annotate_2012 = plot_df_perc[2012].to_dict()
vals_to_annotate_2012
```

```python
## Saving values to annotate at end of plot
vals_to_annotate_2024 = plot_df_perc[2024].to_dict()
vals_to_annotate_2024
```

```python
## Final plot:
with plt.style.context('seaborn'):
    ## Loop through dict to plot each type
    plt.figure(figsize=(12,10))
    for type_, values in data.items():
        plt.plot(years,values, marker='o',label=type_)

    ## Add legend 
    plt.legend(loc='upper left')
    # plt.xlim(2012,2024)

    ## Get rid of frame 
    plt.box(False)
    plt.yticks([]);

    fig = plt.gcf()


    ##Annotate starting values
    for source, val in vals_to_annotate_2012.items():
        ## Annotate value as a % out of 100
        plt.annotate(f"{val:.0f}% - {source}",xy=(2012-.5, val),ha='right')

    ## Annotate final values
    for source, val in vals_to_annotate_2024.items():
        
        ## set fontsize based on %
        if val < .05:
            size = 8
        else: 
            size = 14

        ## Annotate value as a % out of 100
        plt.annotate(f"{val:.0f}% - {source}",xy=(2024+.5, val),fontsize=size)


    plt.axhline(0,xmin=2012,xmax=2020,color='k',ls='-')
    plt.tight_layout()
```

> The annotations are messy, let's use subplots representing the first and last year instead of annotations.

```python
## Get seaborn palette with exact # of colors
palette = sns.color_palette("muted",n_colors=len(plot_df_perc.index))
palette
```

```python
## Making a color mapping dict
color_map = dict(zip(plot_df_perc.index,palette))
color_map
```

```python
## Final plot:

## Specifying widths for gridspec
gridspec = {"width_ratios":[1, 4, 1]}

fig, axes= plt.subplots(figsize=(10,4), ncols=3, sharey=True,
                        gridspec_kw=gridspec,edgecolor=None)

## Plot bar of starting vals
plot_df_perc[[2012]].T.plot(kind='bar',stacked=True,color=color_map.values(),
                            ax=axes[0],legend=None,rot=0)

## Loop through dict to plot each type
for type_, values in data.items():
    c = color_map[type_]
    axes[1].plot(years,values, marker='o',label=type_,color=c)

## Add legend 
axes[1].legend(loc='upper left')

## Plot bar of final vals
plot_df_perc[[2024]].T.plot(kind='bar',stacked=True,color=color_map.values(),
                            ax=axes[2], legend=None,rot=0)

## Get rid of frame 
for ax in axes:
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.yaxis.set_visible(False)
    # ax.set_yticks([])

## Adding back to left figure
axes[0].yaxis.set_visible(True)
axes[0].spines['left'].set_visible(True)
axes[0].set_ylabel('% of Total')

## Final touches
fig.suptitle('Share of U.S. Ad Spend')
plt.tight_layout()
```
